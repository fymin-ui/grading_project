```python
import nbformat

def extract_notebook_content(path):
    nb = nbformat.read(path, as_version=4)

    content = []

    for cell in nb.cells:
        if cell.cell_type == "markdown":
            content.append(cell.source)
        elif cell.cell_type == "code":
            content.append(cell.source)

    return "\n\n".join(content)


notebook_text = extract_notebook_content("prac1-pandas-matplotlib.ipynb")
with open("../extracted_content/student_submission.txt", "w", encoding="utf-8") as f:
    f.write(notebook_text)
```
