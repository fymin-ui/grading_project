```python
import pandas as pd
from openai import OpenAI


df = pd.read_csv("data/ASAP2_train_sourcetexts.csv")

df = df.dropna(subset=["full_text", "assignment"])

df.head(10)

```

```python
# just checking the dataset content and structure

for i in range(5):
    print("Score:", df.iloc[i]["score"])
    print(df.iloc[i]["full_text"][:500])
    print("--------------------------------------------------------")
```

```python
# the prompt func

def build_prompt(assignment, essay):

    return f"""
        You are an academic grader.

        Grade the student essay based on the assignment.

        Assignment:
        {assignment}

        Essay:
        {essay}

        Return this format:

        score: <number>
        
        """


```

```python
# checking the prompt function

essay = df.iloc[0]["full_text"]
assignment = df.iloc[0]["assignment"]

prompt = build_prompt(assignment, essay)

print(prompt)
```

```python
# api request to send the data and the prompt

client = OpenAI(api_key="API_KEY") 

response = client.responses.create(
    model="gpt-4.1",
    input=prompt
)

print(response.output_text)
```
